```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs,make_moons
from sklearn.model_selection import train_test_split
```

```python
X1,y1=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_informative=2,n_clusters_per_class=1,class_sep=2.0,random_state=42)
```

```python
X1
```

```python
y1
```

```python
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.show()
```

```python
X2,y2 = make_blobs(n_samples=1000,centers=2,n_features=2,random_state=42)
```

```python
X2
```

```python
y2
```

```python
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
```

```python
X3,y3 =make_moons(n_samples=1000,noise=0.7)
```

```python
plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.show()
```

```python
X_train,X_test,y_train,y_test =train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
X_train3,X_test3,y_train3,y_test3=train_test_split(X3,y3,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import LinearSVC
```

```python
clf =LinearSVC()
clf.fit(X_train3,y_train3)
```

```python
clf.score(X_train3,y_train3)
```

```python
y_pred=clf.predict(X_test3)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test3,y_pred)
```

```python
clf1 =LinearSVC()
clf1.fit(X_train,y_train)
print(clf1.score(X_train,y_train))
y_pred1=clf1.predict(X_test)
print(accuracy_score(y_test3,y_pred1))
```

```python
X_train2,X_test2,y_train2,y_test2=train_test_split(X2,y2,test_size=0.2,random_state=42)
```

```python
clf2 =LinearSVC()
clf2.fit(X_train2,y_train2)
print(clf2.score(X_train2,y_train2))
y_pred2=clf2.predict(X_test2)
accuracy_score(y_test3,y_pred2)
```
